```python
farben = ["#ff0000", "#ffaa00", "#00aaff", "#888888", "#ffff00", "#000000", "#aaffaa", "#9900cc", "#333333"]
colorStep = 0

farbenForClass = {}
coordsXForClass = {}
coordsYForClass = {}

coordsX = []
coordsY = []
for f in a.features:
    if not f.classifiedAs in farbenForClass:
        farbenForClass[f.classifiedAs] = farben[colorStep]
        coordsXForClass[f.classifiedAs] = []
        coordsYForClass[f.classifiedAs] = []
        colorStep += 1
    
    coord = makeCoordinates2(f)
    coordsXForClass[f.classifiedAs].append(coord[0])
    coordsYForClass[f.classifiedAs].append(coord[1])

# create a new plot with a title and axis labels
p = figure(title="simple circle example", x_axis_label='x = the sum of even indexes', y_axis_label='y = the sum of odd indexes')

# add a line renderer with legend and line thickness
for key in farbenForClass.keys():
    p.triangle(coordsXForClass[key], coordsYForClass[key], size=12, color=farbenForClass[key], legend_label=str(key), line_width=1)

# show the results
show(p)
```

```python
from ImageData import LOOCV, EuclideanDistance, kNearestNeighbour, PipelineManager

l = LOOCV()
l.do(a)

m = PipelineManager()
m.addPipeline(EuclideanDistance())
m.addPipeline(kNearestNeighbour())
m.do(l)
```

```python
len(m.data)
```

```python
countsPerClass = {}
index = 1
for img in m.data:
    if not img.classifiedAs in countsPerClass:
        countsPerClass[img.classifiedAs] = {'positive':0,'negative':0,'index':index}
        index += 1
    countsObj = countsPerClass[img.classifiedAs]
    
    neighbourClass = img.getNeighbourByClass(3)
    
    if neighbourClass == img.classifiedAs:
        countsObj['positive'] += 1
    else:
        countsObj['negative'] += 1
print(countsPerClass)
```

```python
for imgClass in countsPerClass.items():
    print(imgClass)
```

```python
p = figure()

farben = ["#ff0000", "#ffaa00", "#00aaff", "#888888", "#ffff00", "#000000", "#aaffaa", "#9900cc", "#333333"]
for className, imgClass in countsPerClass.items():
    lineWidth = 0.25
    index = int(imgClass['index'])
    xPos = [index - lineWidth/2.0, index + lineWidth/2.0]
    yPos = [imgClass['positive'], imgClass['negative']]
    
    p.vbar(x=xPos, width=lineWidth, bottom=0, top=yPos, color=farben[index - 1], legend_label=className)

show(p)
```
